```python
import re
import glob, os
import itertools
```

```python
# Open file and print it
with open('unique_labels.txt','r') as labels_file:
    file_labels = labels_file.read().splitlines()
file_labels
```

```python
# Get tex files
tex_files = glob.glob('./*.tex')
tex_files.remove('./Thesis-MSc-Abstract-EN-Abstract.tex')
tex_files.remove('./Thesis-MSc-Abstract-EN-KeyWords.tex')
tex_files.remove('./Thesis-MSc-Abstract-PT-PalavrasChave.tex')
tex_files.remove('./Thesis-MSc-Abstract-PT-Resumo.tex')
tex_files.remove('./Thesis-MSc-Acknowledgments.tex')
tex_files.remove('./Thesis-MSc-Acronyms.tex')
tex_files.remove('./Thesis-MSc-Glossary.tex')
```

```python
# Get \eqref and \ref expressions
pattern_1 = r'\\ref\{[^\}]+\}'
pattern_2 = r'\\eqref\{[^\}]+\}'

# Iter through files
iter_matches = []
for file in tex_files:
    with open(file, 'r') as tex_file:
        tex_file_contents = tex_file.read()
        matches_1 = re.findall(pattern_1, tex_file_contents)
        matches_2 = re.findall(pattern_2, tex_file_contents)

    iter_matches.append(matches_1)
    iter_matches.append(matches_2)

# Concatenate all refs
all_matches = list(itertools.chain.from_iterable(iter_matches))

# Remove duplicate refs
all_matches = list(set(all_matches))
```

```python
# Patern to simplify
pattern = r'\\(?:ref|eqref)\{([^}]+)\}'

# Array of used refs/eqrefs
all_references = []

for item in all_matches:
    matches = re.search(pattern, item)
    if matches:
        all_references.append(matches.group(1))
```

```python
# Final loop

unused_labels = []

for label in file_labels:
    if label not in all_references:
        unused_labels.append(label)
```
